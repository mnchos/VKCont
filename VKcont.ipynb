```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train=pd.read_csv('train_df.csv', sep=',')
test=pd.read_csv('test_df.csv', sep=',')
train.head()
```

```python
train_gr=train.groupby('search_id').mean()
test_ge=test.groupby('search_id').mean()
```

```python
X_train=train_gr.drop(['target'],axis=1)
y_train=train_gr['target']
X_test = test_ge.drop([ 'target'], axis=1)
y_test = test_ge['target']
```

```python
import xgboost as xgb
import numpy as np
from sklearn.metrics import make_scorer
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import ndcg_score

def ndcg(y_true, y_pred):
    y_true_binary = np.asarray(y_true)
    y_pred_binary = np.asarray(y_pred)
    return ndcg_score([y_true_binary], [y_pred_binary], k=len(y_true))

dtrain = xgb.DMatrix(X_train, label=y_train)

param_grid = {
    'max_depth': [3, 5, 7],
    'learning_rate': [0.05, 0.1],
    'colsample_bytree': [0.8, 0.9, 1.0],
    'n_estimators': [100, 200, 300]
}

model = xgb.XGBRegressor(objective='reg:squarederror')

ndcg_scorer = make_scorer(ndcg)

grid_search = GridSearchCV(estimator=model, param_grid=param_grid, scoring=ndcg_scorer, cv=5)

grid_search.fit(X_train, y_train)

print("Лучшие параметры:", grid_search.best_params_)

best_model = grid_search.best_estimator_

y_pred_test = best_model.predict(X_test)

ndcg_test = ndcg(y_test, y_pred_test)
print(f"NDCG на тестовых данных: {ndcg_test}")
```
